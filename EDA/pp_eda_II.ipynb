**EJERCICIO 2**

Explorad la columna de country y reportad los resultados solo para aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['country'], thresh= 70, other_label= 'otros paises').round(3) #Filtramos los paises que representan el 70% de registros
```

    Vemos que los paises donde hay mas registros de ataques son USA y Australia, los cuales suman el 65% de todo el conjunto de datos.

---

**EJERCICIO 3**

Explorad la columna country en relación al sexo, filtrando de nuevo por aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['sexo', 'country'], thresh= 70, other_label= 'otros paises').round(3)
```

    Al filtrar el 70% de los casos vemos que nos devuelve en country nuevamente Usa, Australia y agrega South Africa y entendemos que el 70% de  los ataques se dan en hombres por eso solo nos trae M en la columna 'sexo'.

**EJERCICIO 4**

Haced los mismo para la columna activity en relación a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.


```python
df.stb.freq(['activity', 'fatal'], thresh= 60, other_label= 'otras actividades').round(3)
```
